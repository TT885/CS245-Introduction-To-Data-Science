```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
df = pd.read_csv("./adult.txt",names=list('ABCDEFGHIJKLMNO'))
```

```python
# 年龄
# 统计15-90岁的人群分布，以15岁为一组
age = df['A'].values
age_cnt = [0 for _ in range(5)]
age_labels = ['16-30','31-45','46-60','61-75','76-90']
for i in range(5):
    mask = (age>i*15+15) & (age<=i*15+30)
    age_cnt[i] = age[mask].sum()
plt.pie(age_cnt,autopct='%.1f%%')
plt.axis('equal')
plt.legend(age_labels)
plt.savefig('./figures/age.pdf')
plt.show()
```

```python
# 工作类别
work_cls = df['B']
work_cls.value_counts().plot.pie(autopct='%.1f%%')
plt.axis('equal')
plt.savefig('./figures/work_cls.pdf')
plt.show()
```

```python
# 教育程度
edu_lv = df['D']
edu_lv.value_counts().plot.bar()
plt.savefig('./figures/edu_lv.pdf')
plt.show()
```

```python
# 受教育时间
edu_time = df['E']
edu_time.value_counts().plot.bar()
plt.savefig('./figures/edu_time.pdf')
plt.show()
```

```python
# 婚姻状况
marry_stat = df['F']
marry_stat.value_counts().plot.pie(autopct='%.1f%%')
plt.axis('equal')
plt.savefig('./figures/marry_stat.pdf')
plt.show()
```

```python
# 职业
business = df['G']
business.value_counts().plot.bar()
plt.savefig('./figures/business.pdf')
plt.show()
```

```python
# 家庭关系
home_rela = df['H']
home_rela.value_counts().plot.pie(autopct='%.1f%%')
plt.axis('equal')
plt.savefig('./figures/home_rela.pdf')
plt.show()
```

```python
# 种族
race = df['I']
race.value_counts().plot.bar(logy=True)
plt.savefig('./figures/race.pdf')
plt.show()
```

```python
# 职业与每周工作小时数的关系（散点图）
business = df['G']
work_time = df['M']
business = business.values
work_time = work_time.values
# idxs = np.argsort(work_time)
# work_time = work_time[idxs]
# net_earn = net_earn[idxs]
# plt.scatter(work_time, net_earn)
business_cls = np.unique(business)
business_cls = business_cls[business_cls!=' ?']
avg_work_time = []
for obj in business_cls:
    mask = (business==obj)
    avg_work_time.append(np.mean(work_time[mask]))
plt.bar(business_cls,avg_work_time)
plt.xticks(rotation=90)
plt.savefig('./figures/business_worktime.pdf')
plt.show()
```

```python
# 性别和收入的关系（双柱状图）
sex = df['J'].values
income = df['O'].values
sum_male = [np.sum((sex==' Male')&(income==' <=50K.')), np.sum((sex==' Male')&(income==' >50K.'))]
sum_female = [np.sum((sex==' Female')&(income==' <=50K.')), np.sum((sex==' Female')&(income==' >50K.'))]
plt.bar([0,1.5],sum_male,width=0.4,label='Male',fc='b',tick_label=['            <=50K','            >50K'])
plt.bar([0.4,1.9],sum_female,width=0.4,label='Female',fc='r')
plt.legend()
plt.savefig('./figures/sex_income.pdf')
plt.show()
```
