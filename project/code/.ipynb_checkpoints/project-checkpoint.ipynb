# 分类器测试

```python
# 数据转换函数
def data_convert(train_data,test_data,one_hot=False):
    train_data = train_data.copy()
    test_data = test_data.copy()
    str_columns = [1,3,5,6,7,8,9,13]
    train_data_converted = []
    test_data_converted = []
    for column in range(14):
        if column not in str_columns:
            maxval = train_data[column].max()
            minval = train_data[column].min()
            tmptrain = (np.array(train_data[column])-minval)/float(maxval-minval)
            tmptest = (np.array(test_data[column])-minval)/float(maxval-minval)
            train_data_converted.append(np.expand_dims(tmptrain,-1))
            test_data_converted.append(np.expand_dims(tmptest,-1))
            continue
            
        indices = train_data[column].value_counts().index
        for i,index in enumerate(indices):
            train_data.replace(index,i,inplace=True)
            test_data.replace(index,i,inplace=True)
        
        if one_hot:
            enc = OneHotEncoder()
            enc.fit(np.expand_dims(np.array(train_data[column]),-1)) 
            train_data_converted.append(enc.transform(np.expand_dims(np.array(train_data[column]),-1)).toarray())
            test_data_converted.append(enc.transform(np.expand_dims(np.array(test_data[column]),-1)).toarray())
        else:
            maxval = train_data[column].max()
            minval = train_data[column].min()
            tmptrain = (np.array(train_data[column])-minval)/float(maxval-minval)
            tmptest = (np.array(test_data[column])-minval)/float(maxval-minval)
            train_data_converted.append(np.expand_dims(tmptrain,-1))
            test_data_converted.append(np.expand_dims(tmptest,-1))
    
    train_data_converted = np.concatenate(train_data_converted,-1)
    test_data_converted = np.concatenate(test_data_converted,-1)
    
    train_data.replace(' <=50K',0,inplace=True)
    train_data.replace(' >50K',1,inplace=True)
    test_data.replace(' <=50K',0,inplace=True)
    test_data.replace(' >50K',1,inplace=True)
    train_label = np.array(train_data[14],dtype=np.int32)
    test_label = np.array(test_data[14],dtype=np.int32)
    
    return train_data_converted,train_label,test_data_converted,test_label
```

```python
# 训练集和测试集
train_data = pd.read_csv('dataset/adult.data',header=-1,na_values=' ?')
test_data = pd.read_csv('dataset/adult.test',header=-1,na_values=' ?')
test_data = test_data.replace(' <=50K.',' <=50K')
test_data = test_data.replace(' >50K.',' >50K')
train_data = train_data.dropna(how='any')
test_data = test_data.dropna(how='any')
x_train,y_train,x_test,y_test = data_convert(train_data,test_data)
x_train_oh,y_train,x_test_oh,y_test = data_convert(train_data,test_data, one_hot=True)
```

```python
svm_kernel = ['linear','poly','rbf','sigmoid']
penalty = [0.2*x for x in range(1,11)]
svm_penalty_acc = [[] for _ in range(4)]
svm_penalty_training_time = [[] for _ in range(4)]
for C in penalty:
    for i,kernel in enumerate(svm_kernel):
        model = SVC(C=C,kernel=kernel)
        tic = time.time()
        model.fit(x_train,y_train)
        toc = time.time()
        y_pred = model.predict(x_test)
        
        svm_penalty_acc[i].append((y_pred==y_test).mean())
        svm_penalty_training_time[i].append(toc-tic)
        
        print('C = %.1f, kernel = %s' % (C, kernel))
        print('testing accuracy: %f' % (y_pred==y_test).mean())
        print('training time: %f' % (toc-tic))
```

```python
decision_depth = range(1,21)
decision_training_acc = []
decision_testing_acc = []
decision_training_time = []
decision_testing_time = []
for depth in decision_depth:
    model = DecisionTreeClassifier(max_depth=depth)
    tic1 = time.time()
    model.fit(x_train,y_train)
    toc1 = time.time()
    y_train_pred = model.predict(x_train)
    tic2 = time.time()
    y_test_pred = model.predict(x_test)
    toc2 = time.time()
    
    decision_training_acc.append((y_train_pred==y_train).mean())
    decision_testing_acc.append((y_test_pred==y_test).mean())
    decision_training_time.append(toc1-tic1)
    decision_testing_time.append(toc2-tic2)
    
    print('depth = %d' % depth)
    print('training accuracy: %.4f' % ((y_train_pred==y_train).mean()))
    print('testing accuracy: %.4f' % ((y_test_pred==y_test).mean()))
    print('training time: %f' % (toc1-tic1))
    print('testing time: %f' % (toc2-tic2))
    print
```

```python
# KNN
# 算法与时间的关系
KNN_K_values = range(1,21)
KNN_alg = ['ball_tree', 'kd_tree']
# KNN_p = range(1,6)
KNN_training_time_alg = [[] for _ in range(2)]
KNN_testing_time_alg = [[] for _ in range(2)]
KNN_acc_alg = [[] for _ in range(2)]
for K in KNN_K_values:
    for i,alg in enumerate(KNN_alg):
        model = KNeighborsClassifier(n_neighbors=K, algorithm=alg)
        tic1 = time.time()
        model.fit(x_train,y_train)
        toc1 = time.time()
        tic2 = time.time()
        y_pred = model.predict(x_test)
        toc2 = time.time()
        
        KNN_acc_alg[i].append((y_pred==y_test).mean())
        KNN_training_time_alg[i].append(toc1-tic1)
        KNN_testing_time_alg[i].append(toc2-tic2)
        
        print('K = %d， alg = %s' % (K,alg))
        print('accuracy: %.4f' % ((y_pred==y_test).mean()))
        print('training time: %f' % (toc1-tic1))
        print('testing time: %f' % (toc2-tic2))
        print
```

```python
# PCA对比
dim_number = [x for x in range(1,11)]
dim_acc_mlp = []
dim_acc_svm = []
dim_acc_knn = []
dim_acc_decision = []
dim_acc_bayes = []
dim_acc_randomforest = []
dim_training_time_mlp = []
dim_training_time_svm = []
dim_training_time_knn = []
dim_training_time_decision = []
dim_training_time_bayes = []
dim_training_time_randomforest = []

for dim in dim_number:
    pca = PCA(n_components=dim)
    pca.fit(x_train_oh)
    x_train_reduce = pca.transform(x_train_oh)
    x_test_reduce = pca.transform(x_test_oh)
    
    tic = time.time()
    model = SVC()
    model.fit(x_train_reduce, y_train)
    toc = time.time()
    y_pred = model.predict(x_test_reduce)
    dim_acc_svm.append((y_pred==y_test).mean())
    dim_training_time_svm.append(toc-tic)
    print('svm, #dim=%d' % (dim))
    print('training time: %f' % (toc-tic))
    print ('testing accuracy: %f' % (y_pred==y_test).mean())
    print
    
    tic = time.time()
    model = MLPClassifier(hidden_layer_sizes=(32,64,64,16,2),max_iter=20)
    model.fit(x_train_reduce, y_train)
    toc = time.time()
    y_pred = model.predict(x_test_reduce)
    dim_acc_mlp.append((y_pred==y_test).mean())
    dim_training_time_mlp.append(toc-tic)
    print('mlp, #dim=%d' % (dim))
    print('training time: %f' % (toc-tic))
    print ('testing accuracy: %f' % (y_pred==y_test).mean())
    print
    
    tic = time.time()
    model = KNeighborsClassifier()
    model.fit(x_train_reduce, y_train)
    toc = time.time()
    y_pred = model.predict(x_test_reduce)
    dim_acc_knn.append((y_pred==y_test).mean())
    dim_training_time_knn.append(toc-tic)
    print('knn, #dim=%d' % (dim))
    print('training time: %f' % (toc-tic))
    print ('testing accuracy: %f' % (y_pred==y_test).mean())
    print
    
    tic = time.time()
    model = DecisionTreeClassifier()
    model.fit(x_train_reduce, y_train)
    toc = time.time()
    y_pred = model.predict(x_test_reduce)
    dim_acc_decision.append((y_pred==y_test).mean())
    dim_training_time_decision.append(toc-tic)
    print('decision tree, #dim=%d' % (dim))
    print('training time: %f' % (toc-tic))
    print ('testing accuracy: %f' % (y_pred==y_test).mean())
    print
    
    tic = time.time()
    model = GaussianNB()
    model.fit(x_train_reduce, y_train)
    toc = time.time()
    y_pred = model.predict(x_test_reduce)
    dim_acc_bayes.append((y_pred==y_test).mean())
    dim_training_time_bayes.append(toc-tic)
    print('naive bayes, #dim=%d' % (dim))
    print('training time: %f' % (toc-tic))
    print ('testing accuracy: %f' % (y_pred==y_test).mean())
    print
    
    tic = time.time()
    model = RandomForestClassifier(n_estimators=50)
    model.fit(x_train_reduce, y_train)
    toc = time.time()
    y_pred = model.predict(x_test_reduce)
    dim_acc_randomforest.append((y_pred==y_test).mean())
    dim_training_time_randomforest.append(toc-tic)
    print('random forest, #dim=%d' % (dim))
    print('training time: %f' % (toc-tic))
    print ('testing accuracy: %f' % (y_pred==y_test).mean())
    print
```
